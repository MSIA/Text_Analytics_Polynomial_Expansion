```python
import pandas as pd
```

```python
train0=pd.read_csv('train.txt',header=None,sep='=', names=['x','y'])
train0
```

```python
train_size = int(0.7 * len(train0))
val_size = int(0.8 * len(train0))
shuffle_df = train0.sample(frac=1)

# Split your dataset 
train_df = shuffle_df[:train_size]
val_df = shuffle_df[train_size:val_size]
test_df = shuffle_df[val_size:]
```

```python
[train_df.shape,val_df.shape,test_df.shape]
```

```python
train_df.to_csv('train_df3.csv',index=False)
val_df.to_csv('val_df3.csv',index=False)
test_df.to_csv('test_df3.csv',index=False)
```
